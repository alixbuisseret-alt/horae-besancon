```python
import requests
import os
import time

# Liste exacte des folios peints du MS M.293
folios = [
    "001r", "003r", "005r", "007v", "009r",
    "013v", "021r", "029v", "042v", "048r",
    "053r", "058r", "063r", "081r", "103v",
    "112v", "120r", "149r", "160v"
]

manuscrit = "m293"
dossier_img = "2"
output_dir = "morgan_m293_peintures"

os.makedirs(output_dir, exist_ok=True)

for folio in folios:
    nom_fichier = f"{manuscrit}.{folio}.jpg"
    url = f"https://ica.themorgan.org/icaimages/{dossier_img}/{nom_fichier}"
    chemin_local = os.path.join(output_dir, nom_fichier)

    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            with open(chemin_local, "wb") as f:
                f.write(response.content)
            print(f"✅ {nom_fichier}")
        else:
            print(f"❌ {nom_fichier} — code {response.status_code}")
    except Exception as e:
        print(f"⚠️ {nom_fichier} — {e}")

    time.sleep(0.5)

print("\n--- TERMINÉ ---")
print(f"{len(folios)} folios traités")
```

```python
import requests
import os
import time

output_dir = "vienne_onb_1881_peintures"
os.makedirs(output_dir, exist_ok=True)

# URLs exactes extraites de HMML — linkid → image ID (offset +9 confirmé)
miniatures = [
    ("IM00026972", "f027r", "Annonciation"),
    ("IM00026973", "f???v", "Visitation"),
    ("IM00026974", "f???r", "Nativite"),
    ("IM00026975", "f???r", "Annonce_bergers"),
    ("IM00026976", "f???r", "Epiphanie"),
    ("IM00026977", "f???r", "Presentation_temple"),
    ("IM00026978", "f???r", "Fuite_Egypte"),
    ("IM00026979", "f???r", "Vierge_priere"),
    ("IM00026980", "f???r", "Crucifixion"),
    ("IM00026981", "f???r", "Pentecote"),
    ("IM00026982", "f???r", "David_priere"),
    ("IM00026983", "f???r", "Office_morts"),
]

# Headers qui simulent un navigateur humain
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120.0.0.0 Safari/537.36",
    "Referer": "http://18.235.151.129/initials.php",
}

for img_id, folio, sujet in miniatures:
    url = f"http://18.235.151.129/renamed/{img_id}.jpg"
    nom_fichier = f"vienne_1881_{folio}_{sujet}.jpg"
    chemin_local = os.path.join(output_dir, nom_fichier)

    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            with open(chemin_local, "wb") as f:
                f.write(response.content)
            print(f"✅ {nom_fichier}")
        else:
            print(f"❌ {nom_fichier} — code {response.status_code}")
    except Exception as e:
        print(f"⚠️ {nom_fichier} — {e}")

    time.sleep(0.5)

print("\n--- TERMINÉ ---")
```
