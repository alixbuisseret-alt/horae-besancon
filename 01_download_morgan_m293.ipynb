```python
import requests
import os
import time

# Liste exacte des folios peints du MS M.293
folios = [
    "001r", "003r", "005r", "007v", "009r",
    "013v", "021r", "029v", "042v", "048r",
    "053r", "058r", "063r", "081r", "103v",
    "112v", "120r", "149r", "160v"
]

manuscrit = "m293"
dossier_img = "2"
output_dir = "morgan_m293_peintures"

os.makedirs(output_dir, exist_ok=True)

for folio in folios:
    nom_fichier = f"{manuscrit}.{folio}.jpg"
    url = f"https://ica.themorgan.org/icaimages/{dossier_img}/{nom_fichier}"
    chemin_local = os.path.join(output_dir, nom_fichier)

    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            with open(chemin_local, "wb") as f:
                f.write(response.content)
            print(f"✅ {nom_fichier}")
        else:
            print(f"❌ {nom_fichier} — code {response.status_code}")
    except Exception as e:
        print(f"⚠️ {nom_fichier} — {e}")

    time.sleep(0.5)

print("\n--- TERMINÉ ---")
print(f"{len(folios)} folios traités")
```

```python
# Manuscrit : New York, NYPL, Spencer Collection
# 7 enluminures via IIIF Image API v3

ids_images = [
    "425953",
    "425954",
    "425955",
    "425956",
    "425957",
    "435176",
]

output_dir = "nypl_spencer_peintures"
os.makedirs(output_dir, exist_ok=True)

for i, img_id in enumerate(ids_images):
    nom_fichier = f"nypl_spencer_{str(i+1).zfill(2)}_{img_id}.tif"
    url = f"https://iiif.nypl.org/iiif/3/{img_id}/full/max/0/default.tif"
    chemin_local = os.path.join(output_dir, nom_fichier)

    try:
        response = requests.get(url, timeout=900)
        if response.status_code == 200:
            with open(chemin_local, "wb") as f:
                f.write(response.content)
            print(f"✅ {nom_fichier}")
        else:
            print(f"❌ {nom_fichier} — code {response.status_code}")
    except Exception as e:
        print(f"⚠️ {nom_fichier} — {e}")

    time.sleep(1)

print(f"\n--- TERMINÉ — {len(ids_images)} images traitées ---")
```

```python
# Manuscrit : Melbourne, State Library of Victoria, 096/R66Hm
manifest_url = "https://rosetta.slv.vic.gov.au/delivery/iiif/presentation/2.1/IE10525580/manifest.json"
output_dir = "melbourne_slv_096"
os.makedirs(output_dir, exist_ok=True)

response = requests.get(manifest_url)
manifest = response.json()

canvases = manifest["sequences"][0]["canvases"]  # IIIF v2 utilise "sequences"
print(f"Nombre de pages trouvées : {len(canvases)}")

for i, canvas in enumerate(canvases):
    try:
        image_url = canvas["images"][0]["resource"]["@id"]
        label = canvas.get("label", f"page_{i+1}")
        nom_fichier = f"melbourne_{str(i+1).zfill(3)}_{label}.jpg"
        nom_fichier = nom_fichier.replace("/", "-").replace(" ", "_")[:60] + ".jpg"
        chemin_local = os.path.join(output_dir, nom_fichier)

        r = requests.get(image_url, timeout=30, stream=True)
        if r.status_code == 200:
            with open(chemin_local, "wb") as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"✅ {nom_fichier}")
        else:
            print(f"❌ {nom_fichier} — code {r.status_code}")
    except Exception as e:
        print(f"⚠️ Page {i+1} — {e}")

    time.sleep(0.5)

print(f"\n--- TERMINÉ — {len(canvases)} pages traitées ---")
```
