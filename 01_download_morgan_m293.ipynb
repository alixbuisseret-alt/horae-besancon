```python
import requests
import os
import time

# Liste exacte des folios peints du MS M.293
folios = [
    "001r", "003r", "005r", "007v", "009r",
    "013v", "021r", "029v", "042v", "048r",
    "053r", "058r", "063r", "081r", "103v",
    "112v", "120r", "149r", "160v"
]

manuscrit = "m293"
dossier_img = "2"
output_dir = "morgan_m293_peintures"

os.makedirs(output_dir, exist_ok=True)

for folio in folios:
    nom_fichier = f"{manuscrit}.{folio}.jpg"
    url = f"https://ica.themorgan.org/icaimages/{dossier_img}/{nom_fichier}"
    chemin_local = os.path.join(output_dir, nom_fichier)

    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            with open(chemin_local, "wb") as f:
                f.write(response.content)
            print(f"✅ {nom_fichier}")
        else:
            print(f"❌ {nom_fichier} — code {response.status_code}")
    except Exception as e:
        print(f"⚠️ {nom_fichier} — {e}")

    time.sleep(0.5)

print("\n--- TERMINÉ ---")
print(f"{len(folios)} folios traités")
```
